```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
from src.PVC_points import PVC_points
```

```python
def generer_points(n):
    E = rd.random(n)
    F = rd.random(n)

    L = [(float(x), float(y)) for x, y in zip(E, F)]
    return L
```

```python
# Constantes :
NUMBER_OF_POINTS = 10

# Génération des points :
np.set_printoptions(precision=2)
L = generer_points(NUMBER_OF_POINTS)
G = PVC_points()
G.charger_de_liste(L)

plt.title("Les points utilisées dans l'etude")
plt.scatter(np.array(L).T[0], np.array(L).T[1])
```

```python
c_ppp = G.PPP(L[0])
length_ppp = G.longueur(c_ppp)
print(f"La longueur du chemin : {length_ppp:.2f}")

PVC_points.plot_cycle(c_ppp)
```

```python
c_opt_ppp = G.OptPPP(L[0])
length_opt_ppp = G.longueur(c_opt_ppp)

print(f"La longueur du chemin : {length_opt_ppp:.2f}")
PVC_points.plot_cycle(c_opt_ppp, \
                      "Amélioration de la stratégie du point le plus proche")
```

```python
plt.plot(np.array(c_ppp).T[0], np.array(c_ppp).T[1], 'bx-', alpha=0.9, label='PPP')
plt.plot(np.array(c_opt_ppp).T[0], np.array(c_opt_ppp).T[1], 'yx-', alpha=0.6, label='optPPP')
plt.title("Comparaison entre PPP et OptPPP")
plt.legend()
```

```python
c_opt_prim = G.OptPrim()
length_opt_prim = G.longueur(c_opt_prim)

print(f"La longueur du chemin : {length_opt_prim:.2f}")
PVC_points.plot_cycle(c_opt_prim, \
                      f"Stratégie par l’arbre couvrant de poids minimum")
```

```python
c_hds = G.HDS()
length_hds = G.longueur(c_hds)

print(f"La longueur du chemin : {length_hds:.2f}")
PVC_points.plot_cycle(c_hds, \
                      "Recherche exacte HDS")
```
